```python
#import packages

import requests
import statistics
```

```python
# Now, call the Nasdaq API and pull out a small sample of the data (only one day) to get a glimpse
# into the JSON structure that will be returned
url= f'https://data.nasdaq.com/api/v3/datasets/WIKI/TSLA.json?start_date=2017-01-01&end_date=2017-12-31&api_key={API_KEY}'
r=requests.get(url)
json_data = r.json()


#column_names': ['Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Ex-Dividend', 'Split Ratio', 'Adj. Open', 'Adj. High', 'Adj. Low', 'Adj. Close', 'Adj. Volume']
```

```python
TSLA_D = json_data['dataset']
```

```python
TSLA_D['data'][0][2]
```

```python
tesla_price_data=TSLA_D['data']
```

```python
def extract_open(daily_prices):
    return[price[1] for price in daily_prices]
```

```python
tesla_open = extract_open(tesla_price_data)
```

```python
#highest opening price
max(tesla_open)
```

```python
#lowest opening price
min(tesla_open)
```

```python
# define change in any one day (based on High and Low price)
def largest_daily_change(daily_prices):
    return [price[2]-price[3] for price in daily_prices]
```

```python
#list of daily changes
daily_change = largest_daily_change(tesla_price_data)
```

```python
#largest daily change
max(daily_change)
```

```python
#What was the largest change between any two days (based on Closing Price)
#compare difference between consecutive closing prices

#extract list of closing prices
def extract_close(daily_open_close):
    return [price[3] for price in daily_open_close]
```

```python
tesla_close = extract_close(tesla_price_data)
```

```python
closing_price_changes = [close2-close1 for close1, close2 in zip(tesla_close[:-1], tesla_close[1:])]
```

```python
max(closing_price_changes)
```

```python
# What was the average daily trading volume during this year?

#generate ist of average trading volume

def extract_volume(daily_volume):
    return [volume[5] for volume in daily_volume]
```

```python
tesla_daily_volume = extract_volume(tesla_price_data)
```

```python
statistics.mean(tesla_daily_volume)
```

```python
#What was the average daily trading volume during this year

statistics.median(tesla_daily_volume)
```
